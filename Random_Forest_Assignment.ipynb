```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("census-income.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.replace("?",np.nan,inplace = True)
df.dropna(inplace = True)
```

```python
df.rename(columns={"Unnamed: 14":"Annual_income"},inplace = True)
```

```python
df
```

```python
df["Annual_income"] = df["Annual_income"].map({"<=50K":0,">50K":1})
```

```python
x =  df.drop("Annual_income",axis =1)
y = df["Annual_income"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
x_ohe = ohe.fit_transform(x)
x_df = pd.DataFrame(x_ohe,columns = ohe.get_feature_names_out())
```

```python
x_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_df,y,test_size = 0.2,random_state= 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
rf = RandomForestClassifier()
rf1 = rf.fit(x_train,y_train)
```

```python
pred = rf1.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start = 200,stop = 2000,num = 10)]
max_features = ["auto","sqrt"]
max_depth = [int(x) for x in np.linspace(10,110,num = 11)]
max_depth.append(None)
min_samples_split = [2,5,10]
min_samples_leaf = [1,2,4]
bootstrap = [True,False]

random_grid = {"n_estimators":n_estimators,
               "max_features":max_features,
               "max_depth":max_depth,
               "min_samples_split":min_samples_split,
               "min_samples_leaf":min_samples_leaf,
               "bootstrap":bootstrap}


print(random_grid)

```

```python
rand_f = RandomForestClassifier(random_state = 42)
rf_random = RandomizedSearchCV(estimator = rand_f,param_distributions = random_grid,
                               n_iter = 10,scoring = "neg_mean_absolute_error",
                               cv = 3 ,verbose = 2,random_state = 42,
                               n_jobs = 1,return_train_score = True)

rf22 = rf_random.fit(x_train,y_train);
```

```python
rf22.best_params_
```

```python
pred = rf22.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
